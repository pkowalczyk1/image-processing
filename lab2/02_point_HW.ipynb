# Zadanie domowe

W przypadku obrazów w odcieniach szarości pojedynczy piksel z zakresu [0; 255] reprezentowany jest jako 8-bitowa liczba bez znaku.
Pewnym rozszerzeniem analizy sposobu reprezentacji obrazu może być następujący eksperyment.
Załóżmy, że z każdego z 8 bitów możemy stworzyć pojedynczy obraz binarny (ang. _bit-plane slicing_).
Dla obrazka _100zloty.jpg_ (https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/100zloty.jpg) stwórz 8 obrazów, z których każdy powinien zawierać jedną płaszczyznę bitową.
Podpowiedź $-$ warto sprawdzić, jak realizuje się bitowe operacje logiczne.
Zastosowanie takiej dekompozycji obrazu pozwala na analizę ,,ważności'' poszczególnych bitów.
Jest to użyteczne w kwantyzacji, ale także w kompresji.

W drugim etapie zadania proszę spróbować odtworzyć obraz oryginalny z mniejszej liczby obrazów binarnych.
Warto zacząć od dwóch najbardziej znaczących bitów, a później dodawać kolejne.
Należy utworzyć co najmniej trzy wersje zrekonstruowanych obrazów.
Podpowiedź $-$ rekonstrukcja obrazu to mnożenie przez odpowiednią potęgę liczby 2 (przesunięcie bitowe) oraz dodawanie.

```python
import cv2
import os
import requests
import matplotlib.pyplot as plt
import numpy as np

url = 'https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/'
fileName = '100zloty.jpg'
if not os.path.exists(fileName) :
    r = requests.get(url + fileName, allow_redirects=True)
    open(fileName, 'wb').write(r.content)
```

```python
img = cv2.imread(fileName)

plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
def extract_nth_bit(num, bit_pos):
    return (num >> bit_pos) & 1

def nth_bit_plane(img, n):
    img_copy = np.array(img, copy=True)
    for i in range(img_copy.shape[0]):
        for j in range(img_copy.shape[1]):
            img_copy[i][j] = extract_nth_bit(img_copy[i][j], n)
    return img_copy
```

Dekompozycja obrazu

```python
bit_plane_0 = nth_bit_plane(img, 0)

plt.imshow(bit_plane_0.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_1 = nth_bit_plane(img, 1)

plt.imshow(bit_plane_1.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_2 = nth_bit_plane(img, 2)

plt.imshow(bit_plane_2.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_3 = nth_bit_plane(img, 3)

plt.imshow(bit_plane_3.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_4 = nth_bit_plane(img, 4)

plt.imshow(bit_plane_4.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_5 = nth_bit_plane(img, 5)

plt.imshow(bit_plane_5.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_6 = nth_bit_plane(img, 6)

plt.imshow(bit_plane_6.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
bit_plane_7 = nth_bit_plane(img, 7)

plt.imshow(bit_plane_7.astype("bool").astype("uint8"))
plt.xticks([])
plt.yticks([])
plt.show()
```

Odtworzenie obrazu

```python
recovered_1 = cv2.add(bit_plane_7 << 7, bit_plane_6 << 6)

plt.imshow(recovered_1)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
recovered_2 = cv2.add(recovered_1, bit_plane_5 << 5)

plt.imshow(recovered_2)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
recovered_3 = cv2.add(recovered_2, bit_plane_4 << 4)

plt.imshow(recovered_3)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
recovered_6 = cv2.add(recovered_5, bit_plane_1 << 1)

plt.imshow(recovered_6)
plt.xticks([])
plt.yticks([])
plt.show()
```
